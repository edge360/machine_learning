```python
def calculate_distances(whiskey_name, n_neighbors):
    df = pd.read_csv("scotch_mod3.csv", index_col="NAME")
    whiskey_features = df.loc[whiskey_name]
    euclidean_distances = df.apply(lambda x: euclidean(whiskey_features, x), axis=1)
    jaccard_distances = df.apply(lambda x: jaccard(whiskey_features, x), axis=1)
    return euclidean_distances.head(n_neighbors), jaccard_distances.head(n_neighbors)
```

```python
def print_distances(whiskey_name, n_neighbors, euclidean_distances, jaccard_distances):
    print("Closest neighbors to {} (Euclidean distance):".format(whiskey_name))
    print(euclidean_distances.to_string())
    print("Closest neighbors to {} (Jaccard distance):".format(whiskey_name))
    print(jaccard_distances.to_string())
```
